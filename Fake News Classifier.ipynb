```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
dfTrue = pd.read_csv("./True.csv")
dfFake = pd.read_csv("./Fake.csv")
```

```python
dfTrue
```

```python
dfFake
```

```python
dfTrue["isFake"] = 0 # FALSE
dfTrue.head(3)
```

```python
dfFake["isFake"] = 1
dfFake.head(3)
```

```python
newDf = pd.concat([dfTrue, dfFake]).reset_index(drop=True)
newDf
```

```python
newDf.drop(columns=["date"], inplace=True)
```

```python
newDf["combined"] = newDf["title"] + ": " + newDf["text"]
newDf.head()
```

```python
newDf
```

```python
x = newDf["combined"]
y = newDf["isFake"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
vectorization = TfidfVectorizer()
```

```python
x_v_train = vectorization.fit_transform(x_train)
x_v_test = vectorization.transform(x_test)
```

```python
# CREATING THE FIRST MODEL - LOGISTIC REGRESSION CLASSIFIER
LR_model = LogisticRegression()
```

```python
LR_model.fit(x_v_train, y_train)
```

```python
pred_LR_model = LR_model.predict(x_v_test)
```

```python
LR_model.score(x_v_test, y_test)
```

```python
print(classification_report(y_test, pred_LR_model))
```

```python
# CREATING THE SECOND MODEL - Decision Tree Classifier
DT_model = DecisionTreeClassifier()
```

```python
DT_model.fit(x_v_train, y_train)
```

```python
pred_DT_model = DT_model.predict(x_v_test)
```

```python
DT_model.score(x_v_test, y_test)
```

```python
print(classification_report(y_test, pred_DT_model))
```
